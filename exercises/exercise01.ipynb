# Exercise 1.3
Simulate timeseries according to the following models and use a time series plot to decide whether or not these processes are stationary.

a) $Y1: \quad Y_t = E_t â 0.5 \cdot E_{t â 1} \quad , \text{where} E_t \sim N(0, 1) \text{i.i.d.} E_0 = 0$

```r
set.seed(456)

Et <- ts(rnorm(101, 0, 1))
Et[1] <- 0
y1 <- 0 #delete later
for (i in 2:length(Et)) {
    y1[i] <- Et[i] - 0.5 * Et[i - 1]
}
y1 = y1[2:length(y1)]
ts.y1 = ts(y1)

plot(ts.y1)
```

b) $Y2: Y_t = Y_{tâ1} + E_t \quad , \text{where} E_t \sim N(0, 1) \text{i.i.d.} \, Y_0 = 0$

```r
set.seed(456)

Et <- ts(rnorm(101, 0, 1))
y2 <- 0 #delete later
for (i in 2:length(Et)) {
    y2[i] <- y2[i - 1] + Et[i]
}
y2 = y2[2:length(y2)]
ts.y2 = ts(y2)

plot(ts.y2)
```

c) $Y3: Y_t = 0.5 Â· Y_{tâ1} + E_t \quad, \text{where} E_t \sim N(0, 1) i.i.d. Y_0 = 0$

```r
set.seed(456)

Et <- ts(rnorm(101, 0, 1))
y3 <- 0 #delete later
for (i in 2:length(Et)) {
    y3[i] <- 0.5 * y3[i - 1] + Et[i]
}
y3 = y3[2:length(y3)]
ts.y3 = ts(y3)

plot(ts.y3)
```

d) $Y4: Y_t = Y_{tâ1} \cdot E_t \quad, \text{where} E_t â¼ U(0.95, 1.05) \text{i.i.d.} Y_0 = 1$

```r
set.seed(456)

Et <- ts(runif(101, 0.95, 1.05))
y4 <- 1 #delete later
for (i in 2:length(Et)) {
    y4[i] <- y4[i - 1] * Et[i]
}
y4 = y4[2:length(y4)]
ts.y4 = ts(y4)

plot(ts.y4)
```
